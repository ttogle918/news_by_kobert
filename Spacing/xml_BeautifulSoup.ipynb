```python
data_path = '/content/drive/Othercomputers/내 노트북/신문원문PDF'
```

```python
spacing = Spacing()

# 원문과 pykospacing을 적용한 문장을 비교.
# 같다면 True, 다르다면 바뀐 문장 return 
def compare_sent(sent) :
  sent = sent.replace(" ", '')
  kospacing_sent = spacing(sent)
  if kospacing_sent == sent :
    return True
  return kospacing_sent
```

CDATA : Character Data, 문자데이터. parcing 불가능한 범위를 지정한다.


https://stackoverflow.com/questions/2032172/how-can-i-grab-cdata-out-of-beautifulsoup

```python
# page 정보 ( 기사파일명을 dic_article에 담기 )
import os
dic_article = {}
for (path, dir, files) in os.walk(data_path):
  for filename in files :
    if filename[-6:] == '00.xml':   # 페이지 정보
        with open("%s/%s" % (path, filename), "r", encoding="EUC-KR") as f:
          soup = BeautifulSoup(f, "html.parser")
          file_name = soup.find("file_name").text
          articles = soup.findAll("article")
          article_list = []
          for article in articles :
            article_list.append(article.file_name.string)
          dic_article[file_name] = article_list
```

```python
from bs4 import CData
dic_count = {'head' : 0, 'sub' : 0, 'content':0}
for k, li in dic_article.items() :
  print('page : ', k[-7])
  for v in li :
    v = v[:-3] + 'xml'   # 기사 ( 1 ~ 99 )
    with open("%s/%s" % (data_path, v), "r", encoding="utf-8") as f:
      soup = BeautifulSoup(f, "html.parser")
      headline = soup.find("headline")
      subheadline = soup.find("subheadline")
      content = soup.find("datacontent")   # tag 변경 예정

      kospacing_sent = compare_sent(headline.text)
      if kospacing_sent is not True :
        dic_count['head'] += 1
        headline.string.replace_with(kospacing_sent) # replace

      kospacing_sent = compare_sent(subheadline.text)
      if kospacing_sent is not True :
        dic_count['sub'] += 1
        subheadline.string.replace_with(kospacing_sent) # replace
      
      kospacing_sent = compare_sent(content.text[0])
      if kospacing_sent is not True :
        dic_count['content'] += 1
        print(content.text)
        content.string.replace_with(CData(kospacing_sent)) # replace
        print(kospacing_sent)
      print(soup.find("datacontent").text)

      
    break
```
