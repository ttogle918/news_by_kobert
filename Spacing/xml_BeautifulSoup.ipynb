# XML을 BeautifulSoup을 통해 읽고 띄어쓰기 검사하기 ( PyKoSpacing 패키지 사용 )

PyKoSpacing 패키지를 이용하여 제목, 부제목, 본문의 띄어쓰기를 검사

+ 가정 
  1. 신문 내용(배포된 xml 파일)의 맞춤법과 띄어쓰기가 정확하다.
  2. PyKoSpacing은 현재 가장 좋은 띄어쓰기 검사 모델이다.

+ 결과
  1. 띄어쓰기를 제거한 결과

    + 원문과 띄어쓰기한 결과가 다름 : 98,
      + 본문 : 0
      + headline: 75,
      + subheadline : 41
      
  2. 띄어쓰기를 제거하지 않은 결과 ( 원문을 PyKoSpacing )
    + 전체 기사 수 : 157

      + 본문 : 0
      + headline: 148,
      + subheadline : 72

```python
data_path = '/content/drive/Othercomputers/내 노트북/신문원문PDF'
```

```python
spacing = Spacing()

# 원문과 pykospacing을 적용한 문장을 비교.
# 같다면 True, 다르다면 바뀐 문장 return
# 띄어쓰기를 제외한 것, 아닌 것 비교
def compare_sent(sent) :
  # sent = sent.replace(" ", '')
  kospacing_sent = spacing(sent)
  if kospacing_sent == sent :
    return True
  return kospacing_sent
```
