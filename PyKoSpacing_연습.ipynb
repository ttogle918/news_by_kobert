```python
with open("%s/%s" % (data_path, '/train.txt'), encoding="utf-8") as f:
  text = f.readline()
```

```python
text = text.split('.')
len(text)
```

```python
from pykospacing import Spacing
spacing = Spacing()
wrong = []
for i, con in enumerate(text[:100]) :
  kospacing_sent = spacing(con.replace(" ", ''))
  if con != spacing(con) :
    wrong.append(i)
```

```python
for w in wrong :
  print(text[w])
  print(spacing(text[w]), '\n-----------\n')
```
